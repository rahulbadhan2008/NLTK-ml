```python
from nltk.corpus import stopwords
sw=stopwords.words('english')[0:20050]
import string
```

```python
x=tc.iloc[:,19]
y=tc['gender']
```

```python
def tc(x):
    remove_punctuation=[char for char in x if char not in string.punctuation]
    #print(remove_punctuation)
    remove_punctuation=''.join(remove_punctuation)
    #print(remove_punctuation)
    #remove_punctuation.split()
    return[''.join(word) for word in remove_punctuation.split() if word.lower() not in stopwords.words('english')]
```

```python
z=[]
for j in x:
    #print(text_cleaning(j))
    text=" ".join(tc(j))
    z.append(text)
z

```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(z,y,test_size=.25,random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer().fit(x)
print(len(cv.vocabulary_))
```

```python
bow_x_train=cv.fit_transform(x_train)
#print(bow_x)
bow_x_test=cv.transform(x_test)
#print(bow_x_test)
```

```python
X_train=bow_x_train.toarray()
X_test=bow_x_test.toarray()
```

```python
from sklearn.naive_bayes import BernoulliNB
bn=BernoulliNB()
bn.fit(X_train,y_train)
py=bn.predict(X_test)
print(py)
```

```python
from sklearn.metrics import accuracy_score
print("Accurecy Score:",accuracy_score(y_test, py))
```
